```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
#ques1
from sklearn.datasets import load_digits
```

```python
digit=load_digits()
x=digit.data
y=digit.target
x.shape,y.shape
```

```python
#ques2
from sklearn.preprocessing import StandardScaler
x_std= StandardScaler().fit_transform(x)
x_std
```

```python
#ques4
from sklearn.decomposition import PCA
pca=PCA(n_components=8)
pca.fit(x_std)
```

```python
principalComponents=pca.fit_transform(x_std)
principalComponents
```

```python
principalDf=pd.DataFrame(data=principalComponents,columns=['principal component=1','principal component=2','principal component=3','principal component=4','principal component=5','principal component=6','principal component=7','principal component=8'])
principalDf
#new x
```

```python
pca.explained_variance_ratio_
```

```python
#ques3
principalDf_train,principalDf_test,y_train,y_test = train_test_split(principalDf,y,test_size = 0.3)
```

```python
#ques5
reg = LogisticRegression()
reg.fit(principalDf_train,y_train)
```

```python
pred = reg.predict(principalDf_test)
pred
```

```python
#ques6(a)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

```python
#ques6(b)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```

```python
#ques6(c)
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```
